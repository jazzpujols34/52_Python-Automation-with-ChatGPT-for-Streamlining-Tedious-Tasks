# Draw Charts

### Prompt: Plot a linear regression in Python using matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate synthetic data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Fit a linear regression model
lin_reg = LinearRegression()
lin_reg.fit(X, y)

# Predict y values for the regression line
X_new = np.array([[0], [2]])
y_predict = lin_reg.predict(X_new)

# Plot the data points
plt.scatter(X, y, color='blue')
# Plot the regression line
plt.plot(X_new, y_predict, 'r-', linewidth=2, label='Prediction')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

# Send emails using Python

## Prompt: Send an email from "EmailA" to "EmailB" with the subject "Email sent by ChatGPT" and the content  "ChatGPT is Awesome!" using Python

import smtplib
from email.message import EmailMessage

```python
msg = EmailMessage()
msg.set_content("ChatGPT is Awesome!")
msg['Subject'] = "Email sent by ChatGPT"
msg['From'] = "your_email@example.com"
msg['To'] = "recipient_email@example.com"
```

```python
# Connect to the SMTP server
with smtplib.SMTP('smtp.example.com', 587) as server:
    server.starttls()  # Start TLS encryption
    server.login("your_email@example.com", "your_password")  # Login to your email account
    server.send_message(msg)  # Send the email
```

# Web scrapping using ChatGPT

### Prompt: web scrape https://books.toscrape.com/ using Python and beaufitul soup

pip install beautifulsoup4

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://books.toscrape.com/"
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
for book in soup.find_all('article', class_='product_pod'):
    title = book.find('h3').find('a').get('title')
    price = book.find('p', class_='price_color').text
    print(f"Title: {title}\nPrice: {price}\n")
```
